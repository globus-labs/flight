```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

plt.style.use("dark_background")
```

```python
df = pd.read_csv("train_history.csv", index_col=0)
df.head()
```

```python
test_df = pd.read_csv("test_history.csv", index_col=0)
test_df = test_df.rename(columns={"test_loss": "test/loss"})
test_df.head()
```

```python
sns.lineplot(data=df, x="round", y="train/loss")
plt.axhline(test_df.iloc[0]["test/loss"], label="test/loss", color="b")
plt.grid(linestyle=":")
plt.legend()
plt.show()
```
